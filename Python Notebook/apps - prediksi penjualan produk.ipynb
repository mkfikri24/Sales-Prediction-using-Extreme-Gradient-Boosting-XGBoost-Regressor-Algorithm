```python
import joblib
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/MyDrive/Colab Notebooks/prediksi penjualan soto betawi/model - prediksi penjualan produk.sav'
```

```python
model = joblib.load(filename)
```

```python
#Try The Model
data_baru = []

suhu = float(input('Masukkan suhu (°C) : '))
kelembapan = float(input('Masukkan kelembapan (%) : '))
data_baru.append(suhu)
data_baru.append(kelembapan)

data_baru = [data_baru]
cuaca = np.array(data_baru)

new_prediction = model.predict(cuaca)
print(f"Prediksi penjualan harian : {new_prediction[0]}")
```
